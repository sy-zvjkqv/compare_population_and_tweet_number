```python
import pandas as pd
from scipy import signal, stats
```

```python
df_facility = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/overpass/facility_numbers.csv',index_col=0)
df_corr = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/src/basic_analysis/基本統計量.csv',index_col=0)
df_kankou = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/kankou/fix_kankoukyaku.csv',index_col=0)
df_zinkou = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/zinkou/zinkou.csv',index_col=0)
df_distance = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/distance/cal_distance.csv',index_col=0)
```

```python
df_zinkou = df_zinkou.iloc[:,0:30]
```

```python
df_facility = df_facility.iloc[1:9,:]
```

```python
df_corr = df_corr.iloc[1:9,:]
```

```python
df_corr
```

```python
df_merge = df_corr.merge(df_facility,left_index=True,right_index=True)
```

```python
df_merge = df_merge.merge(df_kankou,left_index=True,right_index=True)
```

```python
df_merge = df_merge.merge(df_zinkou,left_index=True,right_index=True)
```

```python
df_merge = df_merge.merge(df_distance,left_index=True,right_index=True)
```

```python
df_merge
```

```python
df_merge.corr()
```

```python
df_merge.columns
```

```python
rank_r = []
for i in df_merge.columns:
    corr, pvalue = stats.spearmanr(df_merge.rank()['相関係数'], df_merge.rank()[i])
    rank_r.append([corr, pvalue])

df_rank = pd.DataFrame(rank_r,index=df_merge.columns, columns=['順位相関係数','p値'])
# df_rank.to_csv('順位相関.csv')

r_r = []
for i in df_merge.columns:
    corr, pvalue = stats.pearsonr(df_merge['相関係数'], df_merge[i])
    r_r.append([corr, pvalue])
df_r = pd.DataFrame(r_r,index=df_merge.columns, columns=['相関係数','p値'])
# df_r.to_csv('相関.csv')
```
