```python
r_r = []
for i in df_merge.columns:
    corr, pvalue = stats.pearsonr(df_merge['相関係数'], df_merge[i])
    r_r.append([corr, pvalue])
df_r = pd.DataFrame(r_r,index=df_merge.columns, columns=['相関係数','p値'])
```
